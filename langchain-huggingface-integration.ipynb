### HuggingFaceEndpoint

```python
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
```

```python
from langchain_core.messages import HumanMessage
```

```python
repo_id = "meta-llama/Llama-3.1-8B-Instruct"
```

```python
llm=HuggingFaceEndpoint(
    repo_id=repo_id,
    max_new_tokens=150,
    temperature=0.7,
    huggingfacehub_api_token=os.getenv("HF_TOKEN"),
    task="conversational",
    provider="auto"
)
```

```python
llm
```

```python
chat=ChatHuggingFace(llm=llm, verbose=True)
```

```python
input="What is Gen AI?"
```

```python
response = chat.invoke([HumanMessage(content=input)])
```

```python
print(response.content)
```

## Model: meta-llama/Llama-3.1-8B-Instruct

```python
from langchain_core.prompts import PromptTemplate
from langchain_classic.chains  import LLMChain
```

```python
llm=HuggingFaceEndpoint(
    repo_id="meta-llama/Llama-3.1-8B-Instruct",
    max_new_tokens=150,
    temperature=0.7,
    huggingfacehub_api_token=os.getenv("HF_TOKEN"), 
    task="conversational",
    provider="auto"
)
```

```python
chat=ChatHuggingFace(llm=llm, verbose=True)
```

```python
template="""
    Question:{question}
    Answer: Let's think step by step.
"""
```

```python
prompt_template=PromptTemplate(template=template,input_variables=["question"])
```

```python
chain = prompt_template | chat
```

```python
response = chain.invoke({"question": "Which Indie band won the internet in 2023?"})
```

```python
print(response.content)
```

```python
from IPython.display import Markdown, display

display(Markdown(response.content))
```

### HuggingFaceEmbeddings - Model: BAAI/bge-small-en

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
hf = HuggingFaceEmbeddings(
    model_name="BAAI/bge-small-en",
    model_kwargs = {'device': "cpu"}, 
    encode_kwargs= {'normalize_embeddings': True}
)
```

```python
hf_embedding = hf.embed_query("Hello HF")
```
